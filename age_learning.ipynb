```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from scipy.stats import pearsonr

# Create a DataFrame with the given data
df = pd.read_excel('./learning_score_visualization/d48_age_score.xlsx')


# Create a scatter plot of age vs. learning score
plt.figure(figsize=(8, 6), dpi=300)
sns.scatterplot(data=df, y='age', x='learning_score', hue='uroa_control', alpha=1, palette=['blue', 'red'])
plt.title('Age vs. Learning Score')
plt.xlabel('Learning Score')
plt.ylabel('Age')
plt.grid(True)


# Display the plot
plt.show()

summary_stats = df.describe()
summary_stats
corr_coefficient, _ = pearsonr(df["age"], df["learning_score"])
corr_coefficient
```
