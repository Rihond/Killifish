# Learning Score
## Introduction:
 This notebook will read the excel files and calculated the difference of Z-score in 6 parameters with 'Day1-Day2', 'Day1-Day3', 'Day1-Day4', 'Day2-Day3', and 'Day2-Day4'. 

```python
import pandas as pd

# Excel File input path
excel_file_path = './d48_learning_score/d48_Learning_Score.xlsx'

# read data
data = pd.read_excel(excel_file_path, sheet_name= 'Z-score')
```

```python

diff_z_score =  {'Name': [],
        'D1_D2_z_dist': [],
        'D1_D3_z_dist': [],
        'D1_D4_z_dist': [],
        'D2_D3_z_dist': [],
        'D2_D4_z_dist': [],
        'D1_D2_z_vel': [],
        'D1_D3_z_vel': [],
        'D1_D4_z_vel': [],
        'D2_D3_z_vel': [],
        'D2_D4_z_vel': [],
        'D1_D2_z_acc': [],
        'D1_D3_z_acc': [],
        'D1_D4_z_acc': [],
        'D2_D3_z_acc': [],
        'D2_D4_z_acc': [],
        'D1_D2_z_head_angle': [],
        'D1_D3_z_head_angle': [],
        'D1_D4_z_head_angle': [],
        'D2_D3_z_head_angle': [],
        'D2_D4_z_head_angle': [],
        'D1_D2_z_y_up': [],
        'D1_D3_z_y_up': [],
        'D1_D4_z_y_up': [],
        'D2_D3_z_y_up': [],
        'D2_D4_z_y_up': [],
        'D1_D2_z_y_mag': [],
        'D1_D3_z_y_mag': [],
        'D1_D4_z_y_mag': [],
        'D2_D3_z_y_mag': [],
        'D2_D4_z_y_mag': [],
        }
diff_z_score = pd.DataFrame(diff_z_score)
diff_z_score
```

```python
# get the distinct fish id
diff_z_score['Name']= data['name'].unique()

m = len(diff_z_score['Name'])
n = len(data['name'])
```

### Calculate the difference of dist

```python
# set i j = 0
i = 0
j = 0
#parameter that you want to calculate
para = data['z_dist']
for i in range(0,n):
    for j in range(0,m):
        if diff_z_score['Name'][j] == data['name'][i]:
            #Get the value of day1
            if data['day'][i] == 1:
                day1 = para[i]
            if data['day'][i] == 2:
                day2 = para[i]
                diff_z_score['D1_D2_z_dist'][j] = abs(para[i]-day1)
            if data['day'][i] == 3:
                diff_z_score['D1_D3_z_dist'][j] = abs(para[i]-day1)
                diff_z_score['D2_D3_z_dist'][j] = abs(para[i]-day2)
            if data['day'][i] == 4:
                diff_z_score['D1_D4_z_dist'][j] = abs(para[i]-day1)
                diff_z_score['D2_D4_z_dist'][j] = abs(para[i]-day2)
    

    
```
